```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z=[8,7,6,5,4]
plt.plot(x, y, label='Line 1')
plt.plot(x, z,label='Line 2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z=[8,7,6,5,4]
m = [1, 4, 9, 16, 25]
plt.plot(x, y, label='Line 1',c='r',linewidth=2)
plt.plot(x, z,label='Line 2',c='b',linewidth=3)
plt.plot(x, m,label='Line 2',c='y',linewidth=1.5)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z=[8,7,6,5,4]
plt.plot(x, y, label='Line 1',c='r',linewidth=2, linestyle='dashdot')
plt.plot(x, z,label='Line 2',c='b',linewidth=3, linestyle='dashed')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z=[8,7,6,5,4]
plt.plot(x, y, label='Line 1',c='r',linewidth=2, marker='o' ,markersize=8)
plt.plot(x, z,label='Line 2',c='b',linewidth=3, marker='s' ,markersize=8)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.legend()
plt.show()
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#',' C++' ]
popularity = [2.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(languages, popularity, color='y')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#',' C++' ]
popularity = [2.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.barh(languages, popularity, color='y')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]
group_names = ['Group1', 'Group2', 'Group3', 'Group4', 'Group5']
num_groups = len(group_names)
bar_width = 0.35
index = np.arange(num_groups)
plt.bar(index, men_means, bar_width, label='Men')
plt.bar(index + bar_width, women_means, bar_width, label='Women')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width / 2, group_names)
plt.legend()
plt.show()
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#',' C++' ]
popularity = [2.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.figure(figsize=(8, 8))
plt.pie(popularity, labels=languages, autopct='%1.1f%%', startangle=140)
plt.title('Popularity of Programming Languages')
plt.show()
```
